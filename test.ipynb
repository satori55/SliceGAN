```python
import pandas as pd
from plotnine import ggplot, aes, geom_segment, geom_point, theme_minimal, labs
from mizani.formatters import percent_format

# 创建示例数据
data = {
    'category': ['A', 'B', 'C', 'D', 'E'],
    'start': [20, 30, 40, 50, 60],
    'end': [70, 60, 50, 40, 30]
}
df = pd.DataFrame(data)

# 创建杠铃图
plot = (ggplot(df, aes(x='start', xend='end', y='category', yend='category'))
        + geom_segment(size=5, color='#D3D3D3')
        + geom_point(aes(x='start'), color='#006D2C', size=4)
        + geom_point(aes(x='end'), color='#B2DF8A', size=4)
        + theme_minimal()
        + labs(title='ggplot2 geom_dumbbell with dot guide'))

# 显示图表
print(plot)
```
